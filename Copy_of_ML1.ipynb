```python
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df, test_df = train_test_split(raw_df, test_size= 0.2, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=0.18, random_state=42)
```

```python
print('train_df.shape :', train_df.shape)
print('val_df.shape :', val_df.shape)
print('test_df.shape :', test_df.shape)
```

```python
train_df
```

```python
input_cols=list(train_df.columns)[2:32]
target_cols = 'diagnosis'
```

```python
print(input_cols)
```

```python
target_cols
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_cols].copy()
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_cols].copy()
```

```python
test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_cols].copy()
```

```python
train_inputs
```

```python
train_targets
```

```python
import numpy as np
```

```python
numerical_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
```

```python
train_inputs[numerical_cols]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(raw_df[numerical_cols])
```

```python
print('Minimum:')
list(scaler.data_min_)
```

```python
print('Maximum:')
list(scaler.data_max_)
```

```python
train_inputs[numerical_cols] = scaler.transform(train_inputs[numerical_cols])
val_inputs[numerical_cols] = scaler.transform(val_inputs[numerical_cols])
test_inputs[numerical_cols] = scaler.transform(test_inputs[numerical_cols])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver = 'liblinear')
```

```python
model.fit(train_inputs[numerical_cols],train_targets)
```

```python
print(model.intercept_)
```

```python
X_train = train_inputs[numerical_cols]
X_val = val_inputs[numerical_cols]
X_test = test_inputs[numerical_cols]
```

```python
train_preds = model.predict(X_train)
```

```python
train_preds
```

```python
train_probs =  model.predict_proba(X_train)
```

```python
train_probs
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(train_targets, train_preds)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(train_targets, train_preds, normalize='true')
```

```python
def predict_and_plot(inputs, targets, name=''):
    preds = model.predict(inputs)
    
    accuracy = accuracy_score(targets, preds)
    print("Accuracy: {:.2f}%".format(accuracy * 100))
    
    cf = confusion_matrix(targets, preds, normalize='true')
    plt.figure()
    sns.heatmap(cf, annot=True)
    plt.xlabel('Prediction')
    plt.ylabel('Target')
    plt.title('{} Confusion Matrix'.format(name));
    
    return preds
```

```python
train_preds = predict_and_plot(X_train, train_targets, 'Training')
```

```python
val_preds = predict_and_plot(X_val, val_targets, 'Validatiaon')
```

```python
test_preds = predict_and_plot(X_test, test_targets, 'Test')
```
